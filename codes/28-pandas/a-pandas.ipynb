## Pandas的运算

### 统计运算

```python
data = pd.read_csv("nba.csv")
players = data.set_index('Name')
players.head()
```

```python
players.describe()
```

### 逻辑运算

```python
df = players['Age'] > 25
df.head()
```

```python
df = players[players['Age'] > 25] 
df.head()
```

### 算数运算

```python
salary = players['Salary']
players['new_salary'] = salary.add(100000)
players.head()
```

```python
players['new_salary'] = salary + 200000
players.head()
```

### 自定义运算

```python
df = players[['Salary']].apply(lambda x: x.max() - x.min())
df
```
