## `ndarray`的创建

```python
# 创建一个长度为10的数组，数组的值都是0。
zeros = np.zeros(10)
print(zeros)
print(zeros.dtype) # 默认的数量类型是浮点型。
```

```python
zeros_two = np.zeros(10, dtype=int) # 创建整形数组。
print(zeros_two)
print(zeros_two.dtype)
```

```python
a = [1, 2, 3]
arr = np.zeros_like(a)
print(arr)
```

```python
# 创建一个3×5的浮点型数组，数组的值都是1。
ones = np.ones(shape=(3, 5), dtype=float)
print(ones)
```

```python
a = [[1, 2, 3], [4, 5, 6]]
arr = np.ones_like(a)
print(arr)
```

## 数据操作

### 改变形状

```python
x = np.arange(10)
print(x)
print(x.ndim)
```

```python
w = x.reshape(2, 5)
print(w)
print(w.ndim)
```

```python
w = x.reshape(10, -1) # -1会根据数据数量和行数自动计算合适的列数。
print(w)
```

### 数据访问

```python
x = np.arange(15).reshape(3, 5)
y = np.arange(10)
print(x)
print(y)
```

```python
# 访问一维数组
print(y)
print(y[0])
print(y[-1])
```

```python
# 访问二维数组
print(x)
print(x[2, 2]) # 使用元组访问，等价于x[(2, 2)]
```

### 数据切片

```python
print(f'原数组={y}')
print(y[2:5])
print(y[:5])
print(y[5:])
print(y[::2])
print(y[::-1])
```

```python
print(x)
print(x[:2, :3]) # 前两行前三列。
print(x[:2, ::2])
```

```python
# 读取第一行
print(x)
print(x[0])
print(x[0, :])
```

```python
# 读取第一列
print(x)
col = x[:, 0]
print(col)
print(col.ndim)
print(col.shape)
```

### 视图

```python
sub_x = x[:2, :3]
print(x)
print(sub_x)
```

```python
sub_x[0, 0] = 100
print(x)
print(sub_x)
```

```python
sub_x = x[:2, :3].copy()
sub_x[0, 0] = 100
print(x)
print(sub_x)
```

## Numpy的运算

### 通用函数

```python
n = 1000000
l = [i for i in range(n)]
```

```python
%%time
double = []
for i in l:
    double.append(2 * i)
print(f"列表长度: {len(double)}")
print("前10个元素:", double[:10])
```

```python
%%time
double = [2 * i for i in l]
print(f"列表长度: {len(double)}")
print("前10个元素:", double[:10])
```

```python
l = np.arange(n)
```

```python
%%time
double = np.array(2 * i for i in l)
```

```python
%%time
double = 2 * l # Numpy支持向量与常数的运算操作
print(double[:10])
```

```python
x = np.arange(1, 16).reshape(3, 5)

print(x)
print(x + 1)
```

```python
print(x)
print(np.sin(x))
```

### 广播机制

```python
a = np.ones((3, 4))
b = np.ones((3, 5))

print(a)
print(b)

print(a + b)
```

### 矩阵运算

```python
a = np.arange(4).reshape(2, 2)
b = np.full((2, 2), 10)
print(a)
print(b)
print(a + b)
```

```python
print(a)
print(b)
print(a.dot(b)) # 矩阵乘法
```

```python
print(a)
print(a.T) # 矩阵的转置
```

```python
inv_a = np.linalg.inv(a)
print(a)
print(inv_a)
```

```python
print(a.dot(inv_a)) # 原局长和逆矩阵矩阵乘法得到单位阵
```

### 聚合运算

```python
big_array = np.random.random(1000000)
%time sum(big_array)
%time np.sum(big_array)
```

```python
x = np.random.normal(0, 1, size=1000000)
print(f'min value index is {np.argmin(x)}') # 获得最小值位置索引
print(f'max value index is {np.argmax(x)}') # 获得最大值位置索引
```

### 数组排序

```python
x = np.arange(16)
np.random.shuffle(x) # 将数组打乱
print(x)

z = np.sort(x) # 排序后返回新数组，x顺序不变
print(z)

print(np.argsort(x)) # 返回排序后数字的索引值，对矩阵同样成立
```

### 高级索引

```python
x = np.arange(16)
print(x)

# 指定特定元素的索引
index = [3, 5, 8]
print(x[index])
```

```python
# 根据原数组索引生成二维矩阵
index = np.array([[0, 2], [1, 3]])
print(x)
print(x[index])
```

```python
w = x.reshape(4, -1)
print(w)

row = np.array([0, 1, 2])
col = np.array([1, 2, 3])
print(w[row, col])
```

```python
print(w[0, col])
print(w[:2, col])
```

```python
col = [True, False, True, True]
print(w[1:3, col])
```

### 布尔运算

```python
x = np.arange(16)
print(x)
print(x < 3)
```

```python
w = x.reshape(4, -1)
print(w)
print(w < 6)
```

```python
print(x)
print(np.sum(x <= 3))  # 统计小于3数据的数量
```

```python
print(w)
print(np.sum((w > 3) & (w < 10))) # 与运算
```

```python
print(x)
print(x[x < 5]) # 索引小于5的值
```
