## 第10题
输出指定字符串A在字符串B中最后出现的位置，如果B中不包含A，返回-1   
从0开始计数  
A="hello"  
B="hi how are you hello world,hello yoyo!"

**学习要点：**  
* 字符串查找功能；  
* while语句使用；

```python
%%time
A = "hello"
B = "hi how are you hello world,hello yoyo!"
last_found = -1
found=B.find(A,0)
while found >=0:
    last_found = found
    found=B.find(A,found+1)
print(last_found)
```

还可以另外一种思路：把A和B倒过来。那么只要找第一个就行了。 
这里有新的技巧：如何将字符串倒转？

```python
%%time
A = "hello"
B = "hi how are you hello world,hello yoyo!"

A = A[::-1]
B = B[::-1]

p = B.find(A)
if p==-1:
    print(p)
else:
    print(len(B)-(p+len(A)))
```

## 第14题
将字符串全部转换为大写或者小写。

```python
def zf_transform(s, t='U'):
    if t.upper() not in ['U', 'D']:
        return s
    if t.upper() == 'U':
        return s.upper()
    else:
        return s.lower()
```

```python
print(zf_transform("hello, My world!",'u'))
print(zf_transform("hello, My world!",'d'))
```

## 第19题
求1000内的水仙花数。  
所谓水仙花数，即一个数等于其各位数的立方和。如：  
$$
153=1^3+5^3+3^3
$$

```python
def istarget(num):
    a = num
    c = 0
    while a > 0:
        b = a%10
        c += b*b*b
        a //= 10
    if c==num:
        return True
    else:
        return False

for i in range(1, 1001):
    if istarget(i):
        print(i)
```

## 第45题
找出列表中出现次数最多的元素

a = [“my”, “skills”, “are”, “poor”, “I”, “am”, “poor”, “I”,  
“need”, “skills”, “more”, “my”, “ability”, “are”,  
“so”, “poor”]

```python
a = ["my","skills","are","poor","I","am","poor","I",
     "need","skills","more","my","ability","are",
     "so","poor"]
```

### 方法1

```python
aset = set(a)
max_freq = 0
max_indx = -1
for x in aset:
    c = a.count(x)
    if c > max_freq:
        max_freq = c
        max_indx = a.index(x)
print(f'出现次数最多的元素是{a[max_indx]}，出现次数是{max_freq}次。')
```

### 方法2

```python
import collections

sa = collections.Counter(a)
target = sa.most_common(1)
print(f'出现次数最多的元素是{target[0][0]}，出现次数是{target[0][1]}次。')
```

## 第47题
给定一个整数数组nums 和一个目标值target ，请你在该数组中找出和为目标值的那两个整数，并返回他们的数组下标。  
你可以假设每种输入只会对应一个答案。但是，数组中同一个元素不能使用两遍。  
示例:  
给定nums=[2，7，11，15]，target=9  
因为nums[0] + nums[1] =2+7 = 9  
所以返回[0， 1]

```python
nums=[2,7,11,15]
target=26
```

```python
def test(nums, target):
    L = len(nums)
    if L < 2:
        return []
    for i in range(L-1):
        for j in range(i+1, L):
            if nums[i] + nums[j] == target:
                return [i,j]
    return []
```

```python
print(test(nums,target))
```

有1、2、3、4数字能组成多少互不相同无重复数的三位数

分别打印这些三位数的组合

```python
d=[1,2,3,4]
for x in d:
    for y in d:
        if x==y: continue
        for z in d:
            if z==y: continue
            if z==x: continue
            print(100*x+10*y+z)
```

## 第69题
a = [11, 2, 33, 1, 5, 88, 3]  
冒泡排序：  
依次比较两个相邻的元素，如果顺序（如从小到大、首字母从A到Z）  
错误就把他们交换过来

```python
a=[11,2,33,1,5,88,3]
```

```python
def bubble(nums):
    al = len(nums)
    n=0
    while n < al:
        for x in range(n+1,al):
            if nums[x]<nums[n]:
                tmp=nums[x]
                nums[x]=nums[n]
                nums[n]=tmp
        n+=1
    return nums
```

```python
bubble(a)
```
